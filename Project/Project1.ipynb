```python
import psycopg2
import pandas as pd
```

```python
def create_database():
    #connect to default database
    conn = psycopg2.connect("host=127.0.0.1 dbname=postgres user=postgres password=1")
    conn.set_session(autocommit=True)
    cur = conn.cursor()
    
    #create new database
    cur.execute("DROP DATABASE IF EXISTS THINH")
    cur.execute("CREATE DATABASE THINH")
    
    #close connection to default database
    conn.close()
    
    #connect to new database
    conn = psycopg2.connect("host=127.0.0.1 dbname=THINH user=postgres password=1")
    cur = conn.cursor()
    
    return cur, conn
    
```
